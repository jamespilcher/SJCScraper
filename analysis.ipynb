# Analyse And Present The Google Docs Data

```python
%load_ext autoreload
%autoreload 2

# Print most common phrase per user
from data import DocsData
google_docs_data = DocsData()
google_docs_data.extend_dataframe()
df = google_docs_data.get_dataframe()
from collections import Counter
# Group by author and aggregate phrases_counter
grouped = df.groupby('author')['phrases_counter'].agg(sum)
print(grouped)
# Find the most common phrases for each author
most_common_phrases = {}
for author, counter in grouped.items():
    most_common_phrases[author] = counter.most_common(15)  # Change 5 to the desired number of common phrases
for author, phrases in most_common_phrases.items():
    print(f'{author}:')
    for phrase, count in phrases:
        print(f'{phrase}: {count}')
    print()
```
